```python
import numpy
import pandas as pd
import requests
import json
```

```python
url = "https://www.thecocktaildb.com/api/json/v1/1/filter.php?c=Cocktail"
user = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'
r = requests.get(url,headers = {'User-agent':user}) # 200 is successful
r
```

```python
myjson = json.loads(r.text)
```

```python
myjson['drinks'][1]
```

```python
myjson['drinks'][1]['strDrink']
```

```python
drink_names = [d['strDrink'] for d in myjson['drinks']]
print(drink_names)
```

```python
pd.json_normalize(myjson, record_path = 'drinks')
```

```python
url = "https://en.wikipedia.org/w/api.php?action=query&list=search&srsearch=Virginia&format=json&srlimit=500"
root = "https://en.wikipedia.org"
end_point  = "/w/api.php"
parameters = {'action':'query',
              'list':'search',
              'srsearch':'Ohio',
              'format':'json',
              'srlimit':500}
r = requests.get(root+end_point,
                headers = {'User-agent':user},
                params = parameters)
myjson = json.loads(r.text)
mydf = pd.json_normalize(myjson, record_path = ['query','search'])
```
