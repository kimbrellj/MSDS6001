# Module 4 Live Coding

```python
import pandas as pd
import numpy as np
import json
from datetime import datetime
import requests
```

```python
root = 'http://aircasting.habitatmap.org'
endpoint ='/api/mobile/sessions.json'
useragent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'
```

```python
time_to = int(datetime(2022,7,5,16,43,0).timestamp())
time_from = int(datetime(2018,7,5,16,43,0).timestamp())
west = -86.788254
south = 29.206719
east = -66.421585
north = 49.415508
```

```python
params = {'time_to':time_to,
          'time_from':time_from,
         'west':west,
         'east':east,
         'north':north,
         'south':south,
         'tags':"",
         "usernames":"",
         'sensor_name':"airbeam2-pm2.5",
         'limit':100,
         'offset':0,
         'measurement_type':"Particulate Matter",
         'unit_symbol':"µg/m³"}
json.dumps(params)
```

```python
stupidparams = {'q':json.dumps(params)}
stupidparams
```

```python
r = requests.get(root + endpoint, headers = {'User-agent':useragent}, params =stupidparams)
r
```

```python
r.text
```

```python
session_df =  pd.json_normalize(json.loads(r.text),record_path='sessions')
```

```python
session_df.columns
```

```python
sessionsID = session_df["streams.AirBeam2-PM2.5.id"]
sessionsID
```

```python
endpoint = '/api/measurements.json'
params = {'stream_ids':sessionsID}
```

```python
r = requests.get(root + endpoint,
                headers = {'User-agent':useragent},
                params = params)
r
```
